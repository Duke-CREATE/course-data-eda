# Course Data Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Data

```python
file_path = 'data/all_minus_med.csv'
df = pd.read_csv(file_path)

pd.set_option('display.max_columns', None)
df.head()
```

## General Content Overview

### Subjects Distribution

```python
# Increase the figure size
plt.figure(figsize=(30, 6))

# Visualize the distribution of subjects in a bar plot
df['Subject'].value_counts().plot(kind='bar')

# Add labels and title
plt.xlabel('Subject')
plt.ylabel('Count')
plt.title('Distribution of Subjects')

# Show the plot
plt.show()
```

## Term Distribution

```python
# plot the 'Term Descr' distribution
plt.figure(figsize=(8, 6))
df['Term Descr'].value_counts().plot(kind='bar')
plt.xlabel('Term Descr')
plt.ylabel('Count')
plt.title('Distribution of Term Descr')
plt.show()
```

### Attribute Distribution

```python
# Get the top 25 values
top_25_values = df['Attribute Formal Desc'].value_counts().head(25)

# Plot the top 25 values in a bar plot
plt.figure(figsize=(15, 6))
top_25_values.plot(kind='bar')
plt.xlabel('Attribute Formal Desc')
plt.ylabel('Count')
plt.title('Top 25 Attribute Formal Desc')
plt.show()
```

### Career Distribution

```python
# plot the career distribution
plt.figure(figsize=(8, 6))
df['Career'].value_counts().plot(kind='bar')
plt.xlabel('Career')
plt.ylabel('Count')
plt.title('Distribution of Career')
plt.show()
```
